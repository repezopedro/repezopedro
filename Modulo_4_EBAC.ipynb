##Aula 2

Modos de abertura de arquivos com o m√©todo 'open':

**r** = read (modo de leitura). Mais seguro!

**w** = write (modo de escrita). Sobrescreve arquivo. Caso o arquivo n√£o exista ele ser√° criado

**a** = Abrir arquivo para acrescentar informa√ß√£o (n√£o sobrescrever)



*   Modo de escrita **w**




```python
#Usando o comando "idades" extra√≠do do arquivo 'banco.csv'
with open(file = 'idades.csv', mode = 'w', encoding = 'utf=8') as fp:
  linha = 'idade' + '\n'
  fp.write(linha)
  for idade in idades:
    linha = str(idade) + '\n'
    fp.write(linha)
```



*   Modo de escrita **a**



```python
with open(file = 'idades.csv', mode = 'a', encoding = 'utf8') as fp:
  for idade in idades:
    linha = str(idade + 100) + '\n'
    fp.write(linha)
```

```python
def escreve_arquivo_csv(nome:str, cabecalho:str, conteudos:list) ->bool:

    try:
        with open(file=nome, mode= 'w', encoding='utf8') as pd:
          linha = cabecalho + '\n'
          pd.write(linha)
          for conteudo in conteudos:
            linha = str(conteudo) + '\n'
            pd.write(linha)

    except Exception as exc:
      print(exc)
      return False

    return True
```

```python
nome = 'idades-funcao.csv'
cabecalho = 'idade'
conteudos = [30, 23, 55, 35, 45, 28, 26, 22, 31, 40]

escreveu_com_sucesso = escreve_arquivo_csv(nome=nome, cabecalho=cabecalho, conteudos=conteudos)
print(escreveu_com_sucesso)
```
